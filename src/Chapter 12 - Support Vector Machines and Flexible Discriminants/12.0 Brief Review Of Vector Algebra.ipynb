## 12.0 Brief Review Of Vector Algebra

Section 4.5 with an example in the Euclidian plane.

### Properties

$X\in\mathbb{R}^p$ and the hyperplane $L\in\mathbb{R}^{p-1}$. $L$ is defined by the equation $f(x)=\beta_0+\beta^{T}x=0$ with $\beta$ a $(p-1)$-vector.

1. $\forall x_0 \in L, \beta^{T}x_0=-\beta_0$  
2. $\forall x_1,x_2 \in L, \beta^{T}(x_1-x_2)=0$ and hence $\beta^{*}=\frac{\beta}{\lVert\beta\rVert}$ is the vector normal to $L$
3. The signed distance of any point $x$ to $L$ is given by:
${\beta^{*}}^{T}(x-x_0)=\frac{f(x)}{\lVert f^{'}(x) \rVert}$ then $f(x)$ gives the signed of the distance

### Illustration

To illustrate these properties we take the follwing example in the Euclidian space:
-  $A=\left\{\begin{matrix}0&3\\\end{matrix}\right\}$ and $B=\left\{\begin{matrix}6&0\\\end{matrix}\right\}$ are two points lying in $L$.
- With these two points we define the equation of the hyperplane (a line in our example) $L$ which is $\beta_0+\beta^Tx=-2+\left(\begin{matrix}0.5&1\\\end{matrix}\right)\left(\begin{matrix}x_1\\x_2\\\end{matrix}\right)=0$.
- $x_a=\left\{\begin{matrix}3&3\\\end{matrix}\right\}$ and $x_b=\left\{\begin{matrix}1.8&0.6\\\end{matrix}\right\}$ are two points not belonging to $L$ and which are symmetrical to $L$.

### Proof

1. Directly with the equation of $L$

2. If $x_1,x_2 \in L$ then $\beta^{T}x_1=-\beta_0$ and $\beta^{T}x_2=-\beta_0$ 
   
   Thus, $\beta^{T}x_1=\beta^{T}x_2$
   
   So, $\beta^{T}(x_1-x_2)=0$
   
3. 
   $={\beta^{*}}^{T}(x-x_0)$
   
   $=\frac{\beta^{T}(x-x_0)}{\lVert\beta\rVert}$
   
   $=\frac{\beta^{T} x-\beta^{T} x_0}{\lVert\beta\rVert}$
   
   $=\frac{\beta^{T} x + \beta_0}{\lVert\beta\rVert}$ because $x_0 \in L$
   
   $=\frac{f(x)}{\lVert f^{'}(x) \rVert}$
